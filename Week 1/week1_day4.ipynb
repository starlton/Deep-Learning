```python
import numpy as np
```

```python
A = np.array([[2, 1],
              [1, 2]])

evals, evecs = np.linalg.eig(A)

print(evals)
print(evecs)
```

### 3. Verify Av = λv for each eigenpair ###

```python
for i in range(len(evals)):
    lambda_ = evals[i]
    vector = evecs[:, i]

    # Calculate Av
    Av = A @ vector
    # Calculate λv
    lambdav = lambda_ * vector

    # Check if Av ≈ λv
    is_close = np.allclose(Av, lambdav)

    print(f"Eigenpair {i+1}: {'Passed' if is_close else 'Failed'}")
    print(f"  Av: {Av}")
    print(f"  λv: {lambdav}")

# 4. Verify all pairs at once (efficient way)
# A @ v matches w * v column-wise due to broadcasting
is_all_correct = np.allclose(A @ evecs, evecs * evals)
print(f"\nAll eigenpairs verified: {is_all_correct}")
```

4. Generate a random (4×4) symmetric matrix: X = np.random.rand(4,4). Get its eigenvalues with eigh. Sort them descending. These sorted eigenvectors are your principal components — print the top 2.

```python
# Random 4x4 Matrix
X = np.random.rand(4,4)

# Symmetric Matrix
S = X.T @ X

# Eigen values of Symmetric
evals, evecs = np.linalg.eigh(S)

#print(evals)

# Print Eigenvalues in Descending order.
sorted_indices = np.argsort(evals)[::-1]
evals_desc = evals[sorted_indices]
evecs_desc = evecs[:, sorted_indices]


print("Eigenvalues in descending order:")
print(evals_desc)

print("Top 2 Principal Components (Eigenvectors)")
print(evecs_desc[:, :2])
```

## **Key Differences between .eig() and .eigh()** ##
### `np.linalg.eig(A)`
*   **Scope:** Works on **any** square matrix.
*   **Output:** Can produce complex/imaginary numbers (may require `.real`).

### `np.linalg.eigh(A)`
*   **Scope:** Optimized specifically for **Symmetric** or **Hermitian** matrices.
*   **Performance:** $\approx$ 2x faster than `.eig()`.
*   **Output:** Guaranteed real numbers (no `.real` needed).

---

### 💡 Why it matters for PCA
*   **Symmetry:** The covariance matrix is always symmetric.
*   **Efficiency:** `eigh()` is faster and more stable.
*   **Convenience:** Results are already sorted in ascending order
