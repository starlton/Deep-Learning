2. Recreate your micrograd example in torch: a=tensor(2., requires_grad=True) etc, f=(a+b)*c, call f.backward(). Confirm a.grad, b.grad, c.grad match what micrograd gave (4, 4, 5).exercise


```python
a = torch.tensor(2., requires_grad=True)
b = torch.tensor(3., requires_grad=True)
c = torch.tensor(4., requires_grad=True)

d = a + b
d.retain_grad()

f = d * c

print(f.item())

f.backward()

a.grad, b.grad, c.grad, d.grad
```

3. Verify a gradient by hand: build a small expression, call backward, check the .grad against your own derivative. Feel that .backward() = your hand-written backward pass.exercise


```python
a = torch.tensor(-4., requires_grad=True)
b = torch.tensor(2., requires_grad=True)
c = a + b ; c.retain_grad()
d = a * b + b**3 ; d.retain_grad()
c += c + 1
c += 1 + c + (-a)
d += d * 2 + (b + a).relu()
d += 3 * d + (b - a).relu()
e = c - d ; e.retain_grad()
f = e**2
f.retain_grad()
g = f / 2.0 ; g.retain_grad()
g += 10.0 / f
print(f'{g.data:.4f}') # prints 24.7041, the outcome of this forward pass
g.backward()

a.grad, b.grad, c.grad, d.grad, e.grad, f.grad, g.grad
```
